```python
# 1) Freight by vendor
freight_sql = """
SELECT VendorNumber, SUM(Freight) AS FreightCost
FROM vendor_invoice
GROUP BY VendorNumber
"""
freight = pd.read_sql(freight_sql, conn)
logging.info("Freight summary built: %d rows", len(freight))

# 2) Purchases by vendor+brand
# NOTE: Volume & ActualPrice come from purchase_prices (alias pp).
purch_sql = """
SELECT
  p.VendorNumber,
  p.VendorName,
  p.Brand,
  p.Description,
  p.PurchasePrice,
  pp.Price  AS ActualPrice,
  pp.Volume,
  SUM(p.Quantity) AS TotalPurchaseQuantity,
  SUM(p.Dollars)  AS TotalPurchaseDollars
FROM purchases p
JOIN purchase_prices pp
  ON p.VendorNumber = pp.VendorNo
 AND p.Brand        = pp.Brand
WHERE p.PurchasePrice > 0
GROUP BY
  p.VendorNumber, p.VendorName, p.Brand, p.Description,
  p.PurchasePrice, pp.Price, pp.Volume
"""
purchases = pd.read_sql(purch_sql, conn)
logging.info("Purchase summary built: %d rows", len(purchases))

# 3) Sales by vendor+brand
sales_sql = """
SELECT
  VendorNo   AS VendorNumber,
  Brand,
  SUM(SalesQuantity) AS TotalSalesQuantity,
  SUM(SalesDollars)  AS TotalSalesDollars,
  SUM(SalesPrice)    AS TotalSalesPrice,
  SUM(ExciseTax)     AS TotalExciseTax
FROM sales
GROUP BY VendorNo, Brand
"""
sales = pd.read_sql(sales_sql, conn)
logging.info("Sales summary built: %d rows", len(sales))

# 4) Merge to one summary
summary = (purchases
           .merge(sales,   on=["VendorNumber","Brand"], how="left")
           .merge(freight, on=["VendorNumber"],         how="left"))

# Clean basics
for c in ["VendorName","Brand","Description"]:
    if c in summary.columns:
        summary[c] = summary[c].astype(str).str.strip()

# force numeric; replace inf with NA then 0
num_cols = [c for c in summary.columns if c not in ["VendorNumber","VendorName","Brand","Description"]]
summary[num_cols] = summary[num_cols].apply(pd.to_numeric, errors="coerce")
summary = summary.replace([pd.NA, float("inf"), -float("inf")], pd.NA).fillna(0)

logging.info("Merged summary shape: %s", summary.shape)
display(summary.head())
```

```python
sql = """
DROP TABLE IF EXISTS vendor_sales_summary;

-- Create table with a CTE: CREATE TABLE ... AS WITH ... SELECT ...
CREATE TABLE vendor_sales_summary AS
WITH
sales_agg AS (
  SELECT p.VendorNumber,
         s.Brand,
         SUM(s.SalesQuantity) AS TotalSalesQuantity,
         SUM(s.SalesDollars)  AS TotalSalesDollars,
         AVG(s.SalesPrice)    AS AvgSalesPrice
  FROM sales s
  JOIN purchases p ON p.InventoryId = s.InventoryId
  GROUP BY p.VendorNumber, s.Brand
),
purch_agg AS (
  SELECT p.VendorNumber,
         p.VendorName,
         p.Brand,
         AVG(pp.PurchasePrice)         AS AvgPurchasePrice,
         AVG(pp.Price)                 AS AvgActualPrice,
         MAX(CAST(pp.Volume AS REAL))  AS Volume
  FROM purchases p
  LEFT JOIN purchase_prices pp
    ON pp.VendorNumber = p.VendorNumber
   AND pp.Brand        = p.Brand
  GROUP BY p.VendorNumber, p.VendorName, p.Brand
),
vi_agg AS (
  SELECT VendorNumber,
         SUM(Quantity) AS TotalPurchaseQuantity,
         SUM(Dollars)  AS TotalPurchaseDollars,
         SUM(Freight)  AS FreightCost
  FROM vendor_invoice
  GROUP BY VendorNumber
)
SELECT
  s.VendorNumber                                   AS VendorNumber,
  COALESCE(TRIM(p.VendorName),'')                  AS VendorName,
  COALESCE(TRIM(s.Brand),'')                       AS Brand,
  COALESCE(p.AvgPurchasePrice,0)                   AS AvgPurchasePrice,
  COALESCE(p.AvgActualPrice,0)                     AS AvgActualPrice,
  COALESCE(p.Volume,0)                             AS Volume,
  COALESCE(s.TotalSalesQuantity,0)                 AS TotalSalesQuantity,
  COALESCE(s.TotalSalesDollars,0)                  AS TotalSalesDollars,
  COALESCE(s.AvgSalesPrice,0)                      AS AvgSalesPrice,
  COALESCE(v.TotalPurchaseQuantity,0)              AS TotalPurchaseQuantity,
  COALESCE(v.TotalPurchaseDollars,0)               AS TotalPurchaseDollars,
  COALESCE(v.FreightCost,0)                        AS FreightCost,

  -- KPIs (in SQL to avoid big pandas ops)
  (COALESCE(s.TotalSalesDollars,0)
   - COALESCE(v.TotalPurchaseDollars,0)
   - COALESCE(v.FreightCost,0))                    AS GrossProfit,

  CASE WHEN COALESCE(s.TotalSalesDollars,0) > 0
       THEN 100.0 * (
              (COALESCE(s.TotalSalesDollars,0)
               - COALESCE(v.TotalPurchaseDollars,0)
               - COALESCE(v.FreightCost,0))
              / COALESCE(s.TotalSalesDollars,0)
            )
       ELSE 0 END                                  AS ProfitMargin,

  CASE WHEN COALESCE(v.TotalPurchaseQuantity,0) > 0
       THEN COALESCE(s.TotalSalesQuantity,0) * 1.0 / v.TotalPurchaseQuantity
       ELSE 0 END                                  AS StockTurnover,

  CASE WHEN COALESCE(v.TotalPurchaseDollars,0) > 0
       THEN COALESCE(s.TotalSalesDollars,0) * 1.0 / v.TotalPurchaseDollars
       ELSE 0 END                                  AS SalesToPurchaseRatio

FROM sales_agg s
LEFT JOIN purch_agg p ON p.VendorNumber = s.VendorNumber AND p.Brand = s.Brand
LEFT JOIN vi_agg v    ON v.VendorNumber = s.VendorNumber
;
"""

t0 = time.time()
conn.executescript(sql)
print("Built vendor_sales_summary in", round((time.time()-t0)/60, 2), "min")
```

```python
# STEP 1  Make vendor-level roll-up (one row per vendor)

import sqlite3, pandas as pd
from pathlib import Path

conn = sqlite3.connect("inventory.db")

sql = """
DROP TABLE IF EXISTS vendor_summary_by_vendor;

CREATE TABLE vendor_summary_by_vendor AS
SELECT
  VendorNumber,
  MIN(TRIM(VendorName))                            AS VendorName,
  COUNT(DISTINCT Brand)                            AS BrandCount,

  -- totals
  SUM(TotalSalesQuantity)                          AS TotalSalesQty,
  SUM(TotalSalesDollars)                           AS TotalSalesDol,
  SUM(TotalPurchaseQuantity)                       AS TotalPurchaseQty,
  SUM(TotalPurchaseDollars)                        AS TotalPurchaseDol,
  SUM(FreightCost)                                 AS FreightCost,

  -- already-computed GP (from brand rows) summed to vendor level
  SUM(GrossProfit)                                 AS GrossProfit,

  -- vendor-level KPIs
  CASE WHEN SUM(TotalSalesDollars) > 0
       THEN 100.0 * SUM(GrossProfit) / SUM(TotalSalesDollars)
       ELSE 0 END                                  AS ProfitMargin,

  CASE WHEN SUM(TotalPurchaseQuantity) > 0
       THEN 1.0 * SUM(TotalSalesQuantity) / SUM(TotalPurchaseQuantity)
       ELSE 0 END                                  AS StockTurnover,

  CASE WHEN SUM(TotalPurchaseDollars) > 0
       THEN 1.0 * SUM(TotalSalesDollars) / SUM(TotalPurchaseDollars)
       ELSE 0 END                                  AS SalesToPurchaseRatio
FROM vendor_sales_summary
GROUP BY VendorNumber;
"""

conn.executescript(sql)

# quick peek
preview = pd.read_sql("""
  SELECT VendorNumber, VendorName, GrossProfit, ProfitMargin, TotalSalesDol, BrandCount
  FROM vendor_summary_by_vendor
  ORDER BY GrossProfit DESC
  LIMIT 10;
""", conn)
display(preview)

# save for dashboards
Path("outputs").mkdir(exist_ok=True)
pd.read_sql("SELECT * FROM vendor_summary_by_vendor;", conn)\
  .to_csv("outputs/vendor_summary_by_vendor.csv", index=False)

conn.close()
```

```python
# === E2E: build summary tables + sanity checks + exports ===
import sqlite3, pandas as pd, time
from pathlib import Path
from IPython.display import display

t0 = time.time()
Path("outputs").mkdir(exist_ok=True)

with sqlite3.connect("inventory.db") as conn:
    # -- light performance knobs (safe) + helpful indexes (idempotent) --
    conn.execute("PRAGMA journal_mode=WAL;")
    conn.execute("PRAGMA temp_store=MEMORY;")
    conn.execute("PRAGMA cache_size=-500000;")  # ~500MB RAM cache if available
    conn.executescript("""
    CREATE INDEX IF NOT EXISTS idx_sales_inventory     ON sales(InventoryId);
    CREATE INDEX IF NOT EXISTS idx_purch_inventory     ON purchases(InventoryId);
    CREATE INDEX IF NOT EXISTS idx_purch_vendor_brand  ON purchases(VendorNumber, Brand);
    CREATE INDEX IF NOT EXISTS idx_pp_vendor_brand     ON purchase_prices(VendorNumber, Brand);
    CREATE INDEX IF NOT EXISTS idx_vi_vendor           ON vendor_invoice(VendorNumber);
    """)

    # ------------------------------------------------------------------
    # 1) BRAND-LEVEL SUMMARY (vendor + brand) -> vendor_sales_summary
    # ------------------------------------------------------------------
    brand_sql = """
    DROP TABLE IF EXISTS vendor_sales_summary;
    CREATE TABLE vendor_sales_summary AS
    WITH
    sales_agg AS (
      SELECT p.VendorNumber,
             s.Brand,
             SUM(s.SalesQuantity) AS TotalSalesQuantity,
             SUM(s.SalesDollars)  AS TotalSalesDollars,
             AVG(s.SalesPrice)    AS AvgSalesPrice
      FROM sales s
      JOIN purchases p ON p.InventoryId = s.InventoryId
      GROUP BY p.VendorNumber, s.Brand
    ),
    purch_agg AS (
      SELECT p.VendorNumber,
             p.VendorName,
             p.Brand,
             AVG(pp.PurchasePrice)        AS AvgPurchasePrice,
             AVG(pp.Price)                AS AvgActualPrice,
             MAX(CAST(pp.Volume AS REAL)) AS Volume
      FROM purchases p
      LEFT JOIN purchase_prices pp
        ON pp.VendorNumber = p.VendorNumber
       AND pp.Brand        = p.Brand
      GROUP BY p.VendorNumber, p.VendorName, p.Brand
    ),
    vi_agg AS (
      SELECT VendorNumber,
             SUM(Quantity) AS TotalPurchaseQuantity,
             SUM(Dollars)  AS TotalPurchaseDollars,
             SUM(Freight)  AS FreightCost
      FROM vendor_invoice
      GROUP BY VendorNumber
    )
    SELECT
      s.VendorNumber                                   AS VendorNumber,
      COALESCE(TRIM(p.VendorName),'')                  AS VendorName,
      COALESCE(TRIM(s.Brand),'')                       AS Brand,
      COALESCE(p.AvgPurchasePrice,0)                   AS AvgPurchasePrice,
      COALESCE(p.AvgActualPrice,0)                     AS AvgActualPrice,
      COALESCE(p.Volume,0)                             AS Volume,
      COALESCE(s.TotalSalesQuantity,0)                 AS TotalSalesQuantity,
      COALESCE(s.TotalSalesDollars,0)                  AS TotalSalesDollars,
      COALESCE(s.AvgSalesPrice,0)                      AS AvgSalesPrice,
      COALESCE(v.TotalPurchaseQuantity,0)              AS TotalPurchaseQuantity,
      COALESCE(v.TotalPurchaseDollars,0)               AS TotalPurchaseDollars,
      COALESCE(v.FreightCost,0)                        AS FreightCost,
      /* KPIs at brand level */
      (COALESCE(s.TotalSalesDollars,0)
       - COALESCE(v.TotalPurchaseDollars,0)
       - COALESCE(v.FreightCost,0))                    AS GrossProfit,
      CASE WHEN COALESCE(s.TotalSalesDollars,0) > 0
           THEN 100.0 * (
                  (COALESCE(s.TotalSalesDollars,0)
                   - COALESCE(v.TotalPurchaseDollars,0)
                   - COALESCE(v.FreightCost,0))
                  / COALESCE(s.TotalSalesDollars,0)
                )
           ELSE 0 END                                  AS ProfitMargin,
      CASE WHEN COALESCE(v.TotalPurchaseQuantity,0) > 0
           THEN COALESCE(s.TotalSalesQuantity,0) * 1.0 / v.TotalPurchaseQuantity
           ELSE 0 END                                  AS StockTurnover,
      CASE WHEN COALESCE(v.TotalPurchaseDollars,0) > 0
           THEN COALESCE(s.TotalSalesDollars,0) * 1.0 / v.TotalPurchaseDollars
           ELSE 0 END                                  AS SalesToPurchaseRatio
    FROM sales_agg s
    LEFT JOIN purch_agg p ON p.VendorNumber = s.VendorNumber AND p.Brand = s.Brand
    LEFT JOIN vi_agg v    ON v.VendorNumber = s.VendorNumber
    ;
    """
    conn.executescript(brand_sql)

    # quick peek
    vs = pd.read_sql("SELECT * FROM vendor_sales_summary LIMIT 10;", conn)
    print("vendor_sales_summary created")
    display(vs)

    # ------------------------------------------------------------------
    # 2) VENDOR-LEVEL ROLLUP (one row per vendor) -> vendor_summary_by_vendor
    # ------------------------------------------------------------------
    rollup_sql = """
    DROP TABLE IF EXISTS vendor_summary_by_vendor;
    CREATE TABLE vendor_summary_by_vendor AS
    SELECT
      VendorNumber,
      MAX(VendorName)                                AS VendorName,
      COUNT(DISTINCT Brand)                          AS BrandCount,
      /* Sums for $ and qty */
      SUM(TotalSalesDollars)                         AS TotalSalesDol,
      SUM(TotalPurchaseDollars)                      AS TotalPurchaseDol,
      SUM(FreightCost)                               AS FreightCost,
      SUM(TotalSalesQuantity)                        AS TotalSalesQty,
      SUM(TotalPurchaseQuantity)                     AS TotalPurchaseQty,
      /* GrossProfit recomputed at vendor level for accuracy */
      SUM(TotalSalesDollars - TotalPurchaseDollars - FreightCost) AS GrossProfit,
      /* Margin percent based on vendor-level totals */
      CASE WHEN SUM(TotalSalesDollars) > 0
           THEN 100.0 * SUM(TotalSalesDollars - TotalPurchaseDollars - FreightCost)
                        / SUM(TotalSalesDollars)
           ELSE 0 END                                 AS ProfitMargin,
      /* Averages for context */
      AVG(AvgSalesPrice)                              AS AvgSalesPrice,
      AVG(AvgPurchasePrice)                           AS AvgPurchasePrice
    FROM vendor_sales_summary
    GROUP BY VendorNumber
    ;
    """
    conn.executescript(rollup_sql)

    preview = pd.read_sql("""
        SELECT VendorNumber, VendorName, GrossProfit, ProfitMargin, TotalSalesDol, BrandCount
        FROM vendor_summary_by_vendor
        ORDER BY GrossProfit DESC
        LIMIT 10;
    """, conn)
    print("vendor_summary_by_vendor created (top 10 by GrossProfit)")
    display(preview)

    # ------------------------------------------------------------------
    # 3) BASIC SANITY CHECKS
    # ------------------------------------------------------------------
    print("\nData quality checks (brand level):")
    dtypes = pd.read_sql("""
      SELECT 'VendorNumber'             AS col, 'INT'  AS expected
      UNION ALL SELECT 'VendorName','TEXT'
      UNION ALL SELECT 'Brand','TEXT'
      UNION ALL SELECT 'TotalSalesDollars','NUM'
      UNION ALL SELECT 'TotalPurchaseDollars','NUM'
      UNION ALL SELECT 'TotalSalesQuantity','NUM'
      UNION ALL SELECT 'TotalPurchaseQuantity','NUM'
      UNION ALL SELECT 'FreightCost','NUM'
    """, conn)
    display(dtypes.head(8))

    nulls = pd.read_sql("""
      SELECT
        SUM(CASE WHEN VendorNumber           IS NULL THEN 1 ELSE 0 END) AS nn_VendorNumber,
        SUM(CASE WHEN VendorName             IS NULL THEN 1 ELSE 0 END) AS nn_VendorName,
        SUM(CASE WHEN Brand                  IS NULL THEN 1 ELSE 0 END) AS nn_Brand,
        SUM(CASE WHEN TotalSalesDollars      IS NULL THEN 1 ELSE 0 END) AS nn_TotalSalesDol,
        SUM(CASE WHEN TotalPurchaseDollars   IS NULL THEN 1 ELSE 0 END) AS nn_TotalPurchaseDol,
        SUM(CASE WHEN TotalSalesQuantity     IS NULL THEN 1 ELSE 0 END) AS nn_TotalSalesQty,
        SUM(CASE WHEN TotalPurchaseQuantity  IS NULL THEN 1 ELSE 0 END) AS nn_TotalPurchaseQty,
        SUM(CASE WHEN FreightCost            IS NULL THEN 1 ELSE 0 END) AS nn_Freight
      FROM vendor_sales_summary;
    """, conn)
    display(nulls)

    negatives = pd.read_sql("""
      SELECT
        SUM(CASE WHEN TotalSalesQuantity    < 0 THEN 1 ELSE 0 END) AS neg_SalesQty,
        SUM(CASE WHEN TotalSalesDollars    < 0 THEN 1 ELSE 0 END) AS neg_SalesDol,
        SUM(CASE WHEN TotalPurchaseQuantity< 0 THEN 1 ELSE 0 END) AS neg_PurchQty,
        SUM(CASE WHEN TotalPurchaseDollars < 0 THEN 1 ELSE 0 END) AS neg_PurchDol,
        SUM(CASE WHEN FreightCost          < 0 THEN 1 ELSE 0 END) AS neg_Freight
      FROM vendor_sales_summary;
    """, conn)
    display(negatives)

    # ------------------------------------------------------------------
    # 4) SMALL INSIGHT TABLES + EXPORTS
    # ------------------------------------------------------------------
    top_vendors = pd.read_sql("""
        SELECT VendorNumber, VendorName, GrossProfit, ProfitMargin, TotalSalesDol
        FROM vendor_summary_by_vendor
        ORDER BY GrossProfit DESC
        LIMIT 10;
    """, conn)

    low_margin = pd.read_sql("""
        SELECT VendorNumber, VendorName, ProfitMargin, TotalSalesDol, GrossProfit
        FROM vendor_summary_by_vendor
        ORDER BY ProfitMargin ASC
        LIMIT 10;
    """, conn)

    display(top_vendors)
    display(low_margin)

    top_vendors.to_csv("outputs/top_vendors_by_grossprofit.csv", index=False)
    low_margin.to_csv("outputs/bottom_vendors_by_margin.csv", index=False)

    # also persist these as tables (optional but handy)
    top_vendors.to_sql("top_vendors_by_grossprofit", conn, if_exists="replace", index=False)
    low_margin.to_sql("bottom_vendors_by_margin",  conn, if_exists="replace", index=False)

print(f"\nDone in {round((time.time()-t0)/60, 2)} min. Files saved in outputs/ and tables saved in inventory.db.")
```

```python
import sqlite3, pandas as pd
from pathlib import Path

conn = sqlite3.connect("inventory.db")

brand_sql = """
DROP TABLE IF EXISTS brand_pricing;

CREATE TABLE brand_pricing AS
SELECT
    VendorNumber,
    VendorName,
    Brand,
    TotalSalesQuantity,
    TotalSalesDollars,
    TotalPurchaseQuantity,
    TotalPurchaseDollars,
    FreightCost,
    -- existing kpis from build
    GrossProfit,
    ProfitMargin,
    SalesToPurchaseRatio,
    StockTurnover,

    -- new diagnostics
    CASE WHEN TotalSalesDollars > 0
         THEN FreightCost * 1.0 / TotalSalesDollars
         ELSE 0 END                       AS FreightPct,

    CASE WHEN TotalPurchaseQuantity > 0
         THEN TotalPurchaseDollars * 1.0 / TotalPurchaseQuantity
         ELSE NULL END                    AS AvgUnitCost,

    CASE WHEN TotalSalesQuantity > 0
         THEN TotalSalesDollars * 1.0 / TotalSalesQuantity
         ELSE NULL END                    AS AvgUnitPrice,

    CASE WHEN TotalSalesQuantity > 0 AND TotalPurchaseQuantity > 0
         THEN (TotalSalesDollars*1.0/TotalSalesQuantity) - (TotalPurchaseDollars*1.0/TotalPurchaseQuantity)
         ELSE NULL END                    AS UnitMargin,

    CASE WHEN TotalSalesQuantity > 0 AND TotalPurchaseQuantity > 0
         THEN ((TotalSalesDollars*1.0/TotalSalesQuantity) - (TotalPurchaseDollars*1.0/TotalPurchaseQuantity))
              / NULLIF((TotalSalesDollars*1.0/TotalSalesQuantity),0)
         ELSE NULL END                    AS UnitMarginPct,

    -- simple, tunable flags
    CASE WHEN ProfitMargin < 10 THEN 1 ELSE 0 END       AS LowMarginFlag,
    CASE WHEN (FreightCost > 0 AND TotalSalesDollars > 0 AND (FreightCost*1.0/TotalSalesDollars) > 0.08)
         THEN 1 ELSE 0 END                               AS HighFreightFlag,
    CASE WHEN (StockTurnover IS NOT NULL AND StockTurnover < 0.5)
         THEN 1 ELSE 0 END                               AS LowTurnoverFlag
FROM vendor_sales_summary;
"""

conn.executescript(brand_sql)

# quick peek & export
df = pd.read_sql("SELECT * FROM brand_pricing ORDER BY ProfitMargin ASC LIMIT 20;", conn)
display(df)

Path("outputs").mkdir(exist_ok=True)
pd.read_sql("SELECT * FROM brand_pricing;", conn).to_csv("outputs/brand_pricing_diagnostics.csv", index=False)

conn.close()
print("brand_pricing created and exported to outputs/brand_pricing_diagnostics.csv")
```

```python
import sqlite3, pandas as pd
from pathlib import Path

# ---- tunable thresholds (adjust later if needed) ----
MIN_SALES_DOLLARS   = 10_000     # ignore tiny brands
MIN_SALES_QTY       = 10
LOW_MARGIN_PCT      = 10.0       # ProfitMargin %
HIGH_FREIGHT_PCT    = 0.08       # 8% of sales
LOW_TURNOVER        = 0.5        # sells < 0.5 of purchases

conn = sqlite3.connect("inventory.db")

sql = f"""
DROP TABLE IF EXISTS underperforming_brands;

CREATE TABLE underperforming_brands AS
SELECT
    VendorNumber,
    VendorName,
    Brand,
    TotalSalesQuantity,
    TotalSalesDollars,
    TotalPurchaseQuantity,
    TotalPurchaseDollars,
    FreightCost,
    GrossProfit,
    ProfitMargin,
    StockTurnover,
    SalesToPurchaseRatio,
    FreightPct,
    AvgUnitCost,
    AvgUnitPrice,
    UnitMargin,
    UnitMarginPct,
    -- reason flags (1=true) to make it clear why it was picked
    CASE WHEN ProfitMargin < {LOW_MARGIN_PCT} THEN 1 ELSE 0 END  AS Reason_LowMargin,
    CASE WHEN FreightPct   > {HIGH_FREIGHT_PCT} THEN 1 ELSE 0 END AS Reason_HighFreight,
    CASE WHEN StockTurnover IS NOT NULL AND StockTurnover < {LOW_TURNOVER} THEN 1 ELSE 0 END AS Reason_LowTurnover
FROM brand_pricing
WHERE
      TotalSalesDollars >= {MIN_SALES_DOLLARS}
  AND TotalSalesQuantity >= {MIN_SALES_QTY}
  AND (
        ProfitMargin < {LOW_MARGIN_PCT}
     OR FreightPct   > {HIGH_FREIGHT_PCT}
     OR (StockTurnover IS NOT NULL AND StockTurnover < {LOW_TURNOVER})
  )
ORDER BY VendorNumber, ProfitMargin ASC, FreightPct DESC;
"""

conn.executescript(sql)

# quick peek & export
Path("outputs").mkdir(exist_ok=True)
df = pd.read_sql("SELECT * FROM underperforming_brands LIMIT 20;", conn)
display(df)

pd.read_sql("SELECT * FROM underperforming_brands;", conn)\
  .to_csv("outputs/underperforming_brands.csv", index=False)

conn.close()
print("underperforming_brands table created + outputs/underperforming_brands.csv exported.")
```

```python
import sqlite3, pandas as pd
from pathlib import Path

conn = sqlite3.connect("inventory.db")

sql = """
DROP TABLE IF EXISTS vendor_risk_summary;

CREATE TABLE vendor_risk_summary AS
WITH bad AS (
  SELECT *
  FROM underperforming_brands
),
v AS (
  -- from earlier step we created this with total vendor sales
  SELECT VendorNumber,
         VendorName,
         TotalSalesDol AS VendorTotalSales
  FROM vendor_summary_by_vendor
)
SELECT
  b.VendorNumber,
  MIN(b.VendorName)                              AS VendorName,
  COUNT(*)                                       AS UnderperformingBrandCount,
  SUM(b.TotalSalesDollars)                       AS SalesAtRisk,
  SUM(b.GrossProfit)                             AS GrossProfit_FromBadBrands,
  SUM(b.FreightCost)                             AS Freight_FromBadBrands,
  AVG(b.ProfitMargin)                            AS AvgMargin_BadBrands,
  MIN(b.ProfitMargin)                            AS MinMargin_BadBrands,
  SUM(CASE WHEN b.Reason_LowMargin   = 1 THEN 1 ELSE 0 END) AS LowMarginBrands,
  SUM(CASE WHEN b.Reason_HighFreight = 1 THEN 1 ELSE 0 END) AS HighFreightBrands,
  SUM(CASE WHEN b.Reason_LowTurnover = 1 THEN 1 ELSE 0 END) AS LowTurnoverBrands,
  v.VendorTotalSales,
  CASE WHEN v.VendorTotalSales > 0
       THEN 1.0 * SUM(b.TotalSalesDollars) / v.VendorTotalSales
       ELSE 0 END                                AS SalesAtRiskPct
FROM bad b
LEFT JOIN v ON v.VendorNumber = b.VendorNumber
GROUP BY b.VendorNumber
ORDER BY SalesAtRisk DESC;
"""

conn.executescript(sql)

# quick peek + export
Path("outputs").mkdir(exist_ok=True)
df = pd.read_sql("SELECT * FROM vendor_risk_summary ORDER BY SalesAtRisk DESC LIMIT 15;", conn)
display(df)

pd.read_sql("SELECT * FROM vendor_risk_summary ORDER BY SalesAtRisk DESC;", conn) \
  .to_csv("outputs/vendor_risk_summary.csv", index=False)

conn.close()
print("vendor_risk_summary table created and exported to outputs/vendor_risk_summary.csv")
```

```python
import sqlite3, pandas as pd
from pathlib import Path

# --- tuning knobs ---
TARGET_MARGIN = 0.35    # aim for 35% margin
MAX_LIFT_PCT  = 0.15    # cap suggested price lift at 15%
MIN_QTY       = 10      # ignore tiny sellers

conn = sqlite3.connect("inventory.db")

sql = f"""
DROP TABLE IF EXISTS brand_pricing_opportunities;

CREATE TABLE brand_pricing_opportunities AS
WITH base AS (
  SELECT
    VendorNumber,
    VendorName,
    Brand,
    TotalSalesQuantity,
    TotalSalesDollars,
    TotalPurchaseDollars,
    FreightCost,
    AvgSalesPrice,
    CASE WHEN TotalSalesDollars > 0
         THEN (TotalSalesDollars - TotalPurchaseDollars - FreightCost) * 1.0
              / TotalSalesDollars
         ELSE 0 END AS CurrentMargin
  FROM vendor_sales_summary
),
filtered AS (
  SELECT *
  FROM base
  WHERE TotalSalesQuantity >= {MIN_QTY}
    AND TotalSalesDollars  >  0
    AND (TotalSalesDollars - TotalPurchaseDollars - FreightCost)
        < ({TARGET_MARGIN}) * TotalSalesDollars
),
need AS (
  SELECT
    *,
    (TotalPurchaseDollars + FreightCost) AS TotalCost,
    CASE WHEN (1.0 - {TARGET_MARGIN}) > 0
         THEN (TotalPurchaseDollars + FreightCost) * 1.0 / (1.0 - {TARGET_MARGIN})
         ELSE NULL END AS RequiredSalesDollars
  FROM filtered
),
calc AS (
  SELECT
    *,
    CASE WHEN RequiredSalesDollars > TotalSalesDollars
         THEN RequiredSalesDollars - TotalSalesDollars
         ELSE 0 END AS UpsideDollars,
    CASE WHEN TotalSalesQuantity > 0
         THEN (CASE WHEN RequiredSalesDollars > TotalSalesDollars
                    THEN (RequiredSalesDollars - TotalSalesDollars)
                    ELSE 0 END) * 1.0 / TotalSalesQuantity
         ELSE 0 END AS NeededPriceDeltaPerUnit
  FROM need
)
SELECT
  VendorNumber,
  VendorName,
  Brand,
  TotalSalesQuantity,
  TotalSalesDollars,
  TotalPurchaseDollars,
  FreightCost,
  AvgSalesPrice,
  CurrentMargin,
  {TARGET_MARGIN} AS TargetMargin,
  UpsideDollars,
  NeededPriceDeltaPerUnit,
  CASE
    WHEN AvgSalesPrice > 0 THEN
      MIN({MAX_LIFT_PCT},
          (NeededPriceDeltaPerUnit / AvgSalesPrice)
      )
    ELSE 0
  END AS SuggestedPriceLiftPct,
  CASE
    WHEN AvgSalesPrice > 0 THEN
      AvgSalesPrice * (1.0 + MIN({MAX_LIFT_PCT},
                                 (NeededPriceDeltaPerUnit / AvgSalesPrice)))
    ELSE AvgSalesPrice
  END AS SuggestedNewAvgPrice
FROM calc
WHERE UpsideDollars > 0
ORDER BY UpsideDollars DESC, CurrentMargin ASC;
"""

conn.executescript(sql)

# preview + export
Path("outputs").mkdir(exist_ok=True)
preview = pd.read_sql("SELECT * FROM brand_pricing_opportunities LIMIT 20;", conn)
display(preview)

pd.read_sql("SELECT * FROM brand_pricing_opportunities;", conn) \
  .to_csv("outputs/brand_pricing_opportunities.csv", index=False)

conn.close()
print("brand_pricing_opportunities created and exported to outputs/brand_pricing_opportunities.csv")
```

```python
import sqlite3, pandas as pd
from pathlib import Path

# ---- tweak thresholds here if you like ----
LOW_MARGIN_PCT      = 0.20   # flag if profit margin < 20%
HIGH_FREIGHT_SHARE  = 0.08   # flag if freight / sales dollars > 8%
LOW_STOCK_TURN      = 0.50   # flag if sales qty / purchase qty < 0.5
MIN_SALES_DOLLARS   = 1000   # ignore tiny vendors

conn = sqlite3.connect("inventory.db")

sql = f"""
DROP TABLE IF EXISTS vendor_risk_flags;

CREATE TABLE vendor_risk_flags AS
WITH base AS (
  SELECT
    VendorNumber,
    VendorName,
    SUM(TotalSalesDollars)      AS TotalSalesDollars,
    SUM(TotalSalesQuantity)     AS TotalSalesQuantity,
    SUM(TotalPurchaseDollars)   AS TotalPurchaseDollars,
    SUM(FreightCost)            AS FreightCost
  FROM vendor_sales_summary
  GROUP BY VendorNumber, VendorName
),
kpi AS (
  SELECT
    *,
    CASE WHEN TotalSalesDollars>0
         THEN (TotalSalesDollars - TotalPurchaseDollars - FreightCost)*1.0 / TotalSalesDollars
         ELSE 0 END AS ProfitMargin,
    CASE WHEN TotalPurchaseDollars>0
         THEN TotalSalesDollars*1.0 / TotalPurchaseDollars
         ELSE 0 END AS SalesToPurchaseRatio,
    CASE WHEN TotalSalesDollars>0
         THEN FreightCost*1.0 / TotalSalesDollars
         ELSE 0 END AS FreightShare,
    CASE WHEN TotalPurchaseQuantity>0 THEN
         TotalSalesQuantity*1.0 / TotalPurchaseQuantity
         ELSE 0 END AS StockTurnover
  FROM (
    SELECT b.*,
           -- derive purchase qty from brand-level table by summing (when available)
           COALESCE((
             SELECT SUM(TotalPurchaseQuantity)
             FROM vendor_sales_summary v
             WHERE v.VendorNumber=b.VendorNumber
           ), 0) AS TotalPurchaseQuantity
    FROM base b
  )
)
SELECT
  VendorNumber,
  VendorName,
  TotalSalesDollars,
  TotalSalesQuantity,
  TotalPurchaseDollars,
  FreightCost,
  ROUND(ProfitMargin*100,2)        AS ProfitMarginPct,
  ROUND(FreightShare*100,2)        AS FreightSharePct,
  ROUND(SalesToPurchaseRatio,2)    AS SalesToPurchaseRatio,
  ROUND(StockTurnover,2)           AS StockTurnover,
  CASE WHEN ProfitMargin < {LOW_MARGIN_PCT} THEN 1 ELSE 0 END AS LowMarginVendor,
  CASE WHEN FreightShare > {HIGH_FREIGHT_SHARE} THEN 1 ELSE 0 END AS HighFreightShare,
  CASE WHEN (TotalSalesDollars - TotalPurchaseDollars - FreightCost) < 0 THEN 1 ELSE 0 END AS NegativeGrossProfit,
  CASE WHEN StockTurnover < {LOW_STOCK_TURN} THEN 1 ELSE 0 END AS LowStockTurnover
FROM kpi
WHERE TotalSalesDollars >= {MIN_SALES_DOLLARS}
ORDER BY ProfitMargin ASC, FreightShare DESC;
"""

conn.executescript(sql)

# preview + export
Path("outputs").mkdir(exist_ok=True)
preview = pd.read_sql("SELECT * FROM vendor_risk_flags LIMIT 20;", conn)
display(preview)

pd.read_sql("SELECT * FROM vendor_risk_flags;", conn) \
  .to_csv("outputs/vendor_risk_flags.csv", index=False)

conn.close()
print("vendor_risk_flags created and exported to outputs/vendor_risk_flags.csv")
```

```python
import sqlite3, pandas as pd
from pathlib import Path

TARGET_MARGIN = 0.35  # change if your target margin is different

conn = sqlite3.connect("inventory.db")

sql = f"""
-- Recreate pricing opportunities with canonical column names
DROP TABLE IF EXISTS brand_pricing_opportunities;
CREATE TABLE brand_pricing_opportunities AS
SELECT
  s.VendorNumber,
  s.VendorName,
  s.Brand,
  s.TotalSalesQuantity,
  s.TotalSalesDollars,
  s.TotalPurchaseDollars,
  s.FreightCost,
  s.AvgSalesPrice,
  CASE 
    WHEN s.TotalSalesDollars > 0 
    THEN (s.TotalSalesDollars - s.TotalPurchaseDollars - s.FreightCost) * 1.0 / s.TotalSalesDollars
    ELSE 0
  END                         AS CurrentMargin,
  CAST({TARGET_MARGIN} AS REAL) AS TargetMargin,
  CASE 
    WHEN s.TotalSalesDollars > 0 
     AND (s.TotalSalesDollars - s.TotalPurchaseDollars - s.FreightCost) * 1.0 / s.TotalSalesDollars < {TARGET_MARGIN}
    THEN (s.TotalPurchaseDollars + s.FreightCost) * 1.0 / (1.0 - {TARGET_MARGIN})
    ELSE NULL
  END                         AS RequiredSalesDollars,
  CASE 
    WHEN s.TotalSalesDollars > 0 
     AND (s.TotalSalesDollars - s.TotalPurchaseDollars - s.FreightCost) * 1.0 / s.TotalSalesDollars < {TARGET_MARGIN}
    THEN ( (s.TotalPurchaseDollars + s.FreightCost) * 1.0 / (1.0 - {TARGET_MARGIN}) ) - s.TotalSalesDollars
    ELSE NULL
  END                         AS UpsideDollars,
  CASE 
    WHEN s.TotalSalesQuantity > 0
     AND s.TotalSalesDollars > 0
     AND (s.TotalSalesDollars - s.TotalPurchaseDollars - s.FreightCost) * 1.0 / s.TotalSalesDollars < {TARGET_MARGIN}
    THEN ( ( (s.TotalPurchaseDollars + s.FreightCost) * 1.0 / (1.0 - {TARGET_MARGIN}) ) - s.TotalSalesDollars ) * 1.0 / s.TotalSalesQuantity
    ELSE NULL
  END                         AS NeededPriceDeltaPerUnit
FROM vendor_sales_summary s
;

-- Recreate final tables
DROP TABLE IF EXISTS final_vendor;
DROP TABLE IF EXISTS final_brand;

-- final_vendor: one row per vendor
CREATE TABLE final_vendor AS
SELECT
  v.VendorNumber,
  v.VendorName,
  t.TotalSalesDollars,
  t.TotalSalesQuantity,
  t.TotalPurchaseDollars,
  t.FreightCost,
  v.GrossProfit,
  v.ProfitMargin,
  COALESCE(v.TotalSalesDol, 0)  AS TotalSalesDol_in_summary,
  COALESCE(s.BrandCount, 0)     AS BrandCount,
  COALESCE(r.neg_SalesQty, 0)   AS neg_SalesQty,
  COALESCE(r.neg_SalesDol, 0)   AS neg_SalesDol,
  COALESCE(r.neg_PurchQty, 0)   AS neg_PurchQty,
  COALESCE(r.neg_PurchDol, 0)   AS neg_PurchDol,
  COALESCE(r.neg_Freight, 0)    AS neg_Freight
FROM vendor_summary_by_vendor v
LEFT JOIN (
    SELECT
      VendorNumber,
      SUM(TotalSalesQuantity)   AS TotalSalesQuantity,
      SUM(TotalSalesDollars)    AS TotalSalesDollars,
      SUM(TotalPurchaseDollars) AS TotalPurchaseDollars,
      SUM(FreightCost)          AS FreightCost
    FROM vendor_sales_summary
    GROUP BY VendorNumber
) t
  ON t.VendorNumber = v.VendorNumber
LEFT JOIN spend_by_vendor   s ON s.VendorNumber = v.VendorNumber
LEFT JOIN vendor_risk_flags r ON r.VendorNumber = v.VendorNumber
;

-- final_brand: vendor+brand with pricing opportunity fields
CREATE TABLE final_brand AS
SELECT
  s.VendorNumber,
  s.VendorName,
  s.Brand,
  s.TotalSalesQuantity,
  s.TotalSalesDollars,
  s.TotalPurchaseDollars,
  s.FreightCost,
  s.GrossProfit,
  s.ProfitMargin         AS CurrentMargin,
  b.TargetMargin         AS TargetMargin,
  b.RequiredSalesDollars,
  b.UpsideDollars,
  b.NeededPriceDeltaPerUnit
FROM vendor_sales_summary s
LEFT JOIN brand_pricing_opportunities b
  ON b.VendorNumber = s.VendorNumber
 AND b.Brand        = s.Brand
;
"""
conn.executescript(sql)

# preview + export
Path("outputs").mkdir(exist_ok=True)

final_vendor_preview = pd.read_sql(
    "SELECT * FROM final_vendor ORDER BY GrossProfit DESC LIMIT 10;", conn)
final_brand_preview = pd.read_sql(
    "SELECT * FROM final_brand ORDER BY COALESCE(UpsideDollars,0) DESC LIMIT 10;", conn)

display(final_vendor_preview)
display(final_brand_preview)

pd.read_sql("SELECT * FROM final_vendor;", conn).to_csv("outputs/final_vendor.csv", index=False)
pd.read_sql("SELECT * FROM final_brand;",  conn).to_csv("outputs/final_brand.csv",  index=False)

conn.close()
print("Built + exported: outputs/final_vendor.csv and outputs/final_brand.csv")
```
